# Preparing the dataframe for analysis 

```python
#reading the excel file
df_skeleton_analysis = pd.read_excel('C:\\Users\\user\\Desktop\\Skeleton_Analysis\\Skeleton_Analysis_42136_RF_W01VideoProcessed2021-12-07T12_11_59.xlsx')
df_skeleton_analysis
```

```python
# remove index 0
df_skeleton_analysis = df_skeleton_analysis.drop(index=0)
# show the dataframe
df_skeleton_analysis
```

```python
# create a dataframe containing only the orientations of the "bones"
df_orientations = df_skeleton_analysis[['left_frontlimb_digitaltip_left_frontlimb_heel_1',
                                        'right_frontlimb_digitaltip_right_frontlimb_heel_3',
                                        'left_hindlimb_digitaltip_left_hindlimb_heel_5',
                                        'right_hindlimb_digitaltip_right_hindlimb_heel_7']]
# show the new dataframe
# IMPORTANT: the orientations are clockwise from a horizontal line and their directions are always heel->digitaltip
df_orientations
```

```python
# change column names ... 
df_orientations.rename(columns = {'left_frontlimb_digitaltip_left_frontlimb_heel_1':'LF_paw', 
                                  'right_frontlimb_digitaltip_right_frontlimb_heel_3':'RF_paw', 
                                  'left_hindlimb_digitaltip_left_hindlimb_heel_5':'LB_paw', 
                                  'right_hindlimb_digitaltip_right_hindlimb_heel_7':'RB_paw'}, inplace = True)
# show the new column names
df_orientations
```

```python
# these orientations have the direction heel->digitaltip
# For easier visualization of results I will invert the direction of the "vectors"
df_orientations_corrected = pd.DataFrame()
for column_name in df_orientations.columns:
    new_orientations=[]
    for orientation in df_orientations[column_name].values.astype(float):
        if orientation > 180:
            orientation-=180
            new_orientations.append(orientation)
        else:
            orientation+=180
            new_orientations.append(orientation)
    df_orientations_corrected[column_name] = new_orientations
# showing dataframe
df_orientations_corrected
```

```python
# showing df_orientations_corrected frame, whose orientations have the direction heel->digitaltip
# attention! in this dataframe the first index is 0
df_orientations_corrected
```

```python
df_orientations_corrected['angle_hindlimbs'] = df_orientations_corrected['LB_paw'].astype(float)-df_orientations_corrected['RB_paw'].astype(float)
```

```python
import dlc2kinematics as dlck
import pandas as pd
import numpy as np
```

```python
# load data
df, bodyparts, scorer = dlck.load_data('F:\\G3\\W01\\42136_RF\\42136_RF_W01VideoProcessed2021-12-07T12_11_59DLC_resnet50_Dystonia_TestApr21shuffle1_500000_filtered.h5')
```

```python
# create a dataframe with the velocity (in x and y) of the 'nose'
df_vel_nose_tailbase = dlck.compute_velocity(df,bodyparts=['nose', 'tailbase'])
```

```python
# show the dataframe 
df_vel_nose_tailbase
```

```python
# change columns names
df_vel_nose_tailbase.columns = ['nose v(x)', 'nose v(y)', 'nose likelihood', 'tailbase v(x)', 'tailbase v(y)', 'tailbase likelihood']
# show the dataframe
df_vel_nose_tailbase
```

```python
import math
```

```python
#convert the values to float
for col in df_vel_nose_tailbase:
    df_vel_nose_tailbase[col] = pd.to_numeric(df_vel_nose_tailbase[col], downcast="float")
```

```python
# create two new columns with v = vx**2 + vy**2, for both the nose and the tailbase

velocity_nose = np.sqrt(df_vel_nose_tailbase['nose v(x)']**2+df_vel_nose_tailbase['nose v(y)']**2)
velocity_tailbase = np.sqrt(df_vel_nose_tailbase['tailbase v(x)']**2+df_vel_nose_tailbase['tailbase v(y)']**2)
df_vel_nose_tailbase['Velocity nose (px/frame)'] = velocity_nose
df_vel_nose_tailbase['Velocity tailbase (px/frame)'] = velocity_tailbase
```

```python
# show the dataframe
df_vel_nose_tailbase
```

```python
# search for NaN's (not a number)
np.sqrt(df_vel_nose_tailbase['nose v(x)']**2+df_vel_nose_tailbase['nose v(y)']**2).isnull().values.any()
```

```python
# check the type of data of the dataframe elements
for col in df_vel_nose_tailbase.columns:
    print(df_vel_nose_tailbase[col].dtype)
```

```python
# define tresholds for gaits
    # standing steel: vel < 5cm/s
    # walking: 5 < vel < 10 cm/s
    # troting: vel > 10 cm/s

# convert from px/frame to cm/s (it will be an estimate, since the video is distorted)
# 40 cm --- 600 px
# 1 cm --- ? px

#ONLY AN ESTIMATE!!! (NOT AN ACURATE CONVERSION)
open_field_length_cm = 40
open_field_length_px = 628
ratio_px_to_cm = open_field_length_px/open_field_length_cm
print(ratio_px_to_cm)

# convert frames to seconds (30 frames = 1 sec)
acquisition_rate_hz = 30 
print(acquisition_rate_hz)

convert = ratio_px_to_cm/acquisition_rate_hz
print(convert)
```

```python
# create a new dataframe with the velocities in cm/sec
df_vel_cms = df_vel_nose_tailbase[['Velocity nose (px/frame)', 'Velocity tailbase (px/frame)']]*convert
df_vel_cms = df_vel_cms.rename(columns = {'Velocity nose (px/frame)':'Velocity nose (cm/s)', 'Velocity tailbase (px/frame)':'Velocity tailbase (cm/s)'})

# show converted dataframe
df_vel_cms
```

```python
# create a column with the time in seconds
df_vel_cms['Time (s)'] = np.arange(len(df_vel_cms))
#.../acquisition_rate_hz = 30 
df_vel_cms['Time (s)'] = df_vel_cms['Time (s)']/acquisition_rate_hz

# now, to ensure that 30 frames correspond to 1 sec...
df_vel_cms.head(31)
```

```python
# get the cell value from the last row
print(str(df_vel_cms.iloc[-1,2]) + ' ' + 'seconds')
# or print (df_vel_cms['Time (s)'].iloc[-1])

# convert to min
import datetime
str('The duration of the video in h:min:sec is:') + str(' ') + str(datetime.timedelta(seconds=df_vel_cms.iloc[-1,2]))
```

```python
# create a new column with the time in minutes
df_vel_cms['Time (min)'] = np.arange(len(df_vel_cms))
#.../acquisition_rate_hz = 30 
df_vel_cms['Time (min)'] = df_vel_cms['Time (min)']/(acquisition_rate_hz*60)

# show the new column
df_vel_cms
```

```python
# see max velocity
df_vel_cms.sort_values(by='Velocity tailbase (cm/s)')
```

```python
# showing the dataframe
df_vel_cms
```

```python
# count frames where velocity is in each of the intevals

# the mice is considered to be standing still if v<=1cm/s
upperl_standing_stil = 1
# the other gaits were established using the mean velocity at each gait and an estimated deviation
# from the mean
mean_vel_walking = 5
mean_vel_trot = 15
estimate_deviation = 2

# creating variables for the velocity limits of the gaits
lowerl_walking = mean_vel_walking - estimate_deviation
upperl_walking = mean_vel_walking + estimate_deviation
lowerl_trot = mean_vel_trot - estimate_deviation
upperl_trot = mean_vel_trot + estimate_deviation

# calculating the time (in frames) spent in each interval of velocities (in a specific video
# ...for a specific mice)
bins = [0, upperl_standing_stil, lowerl_walking, upperl_walking, lowerl_trot, upperl_trot, 500]
df_percentage_gait = df_vel_cms['Velocity tailbase (cm/s)'].value_counts(bins=bins, sort=False) 
df_percentage_gait
```

# Corrections (outward, inward, and transition from 0 to 360) 

## Option 1 

```python
# only corrects for the cases where RB > LB AND the paws are oriented outwards
df_angles = pd.DataFrame()

angle_hindlimb_paws=[]
for LB, RB in zip(df_standing_still['LB_paw'].values.astype(float), df_standing_still['RB_paw'].values.astype(float)):
    if RB > LB:
        angle=LB+(360-RB)
        angle_hindlimb_paws.append(angle)
    else:
        angle=LB-RB
        angle_hindlimb_paws.append(angle)
df_angles[''] = angle_hindlimb_paws

# showing dataframe
df_angles
```

```python
# plot the kernel density estimation of the hind paw angles while standing still
import numpy as np
import matplotlib.pyplot as plt

# only two visible bins
df_angles.plot(kind='kde', figsize=(20,8), xlim=(-360,360))

# plot everything in the same figure
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df_angles.astype(float, errors='raise').plot(kind='line',
                                                                     xlabel='Frame number', 
                                                                     ylabel='Angle of the front paws (º)',
                                                                     figsize=(20,4))
plt.show()   
```

```python
# get the mean amplitude
mean_amplitude = str(round(df_angles.mean(), 2))
print('The mean angle of the hind paws while standing still is: ', mean_amplitude)
```

```python
# there is a peak close to -300 which is most likely related to the transition between 0 to 360 degrees while the hind
# paws are oriented outwards

# In this respect, I will try to only perform an operation in angles below -200

# only corrects for the cases where RB > LB AND the paws are oriented outwards
df_angles_2 = pd.DataFrame()

angle_hindlimb_paws=[]
for LB, RB in zip(df_standing_still['LB_paw'].values.astype(float), df_standing_still['RB_paw'].values.astype(float)):
    if LB-RB < -200:
        angle=LB+(360-RB)
        angle_hindlimb_paws.append(angle)
    else:
        angle=LB-RB
        angle_hindlimb_paws.append(angle)
df_angles_2[''] = angle_hindlimb_paws

# showing dataframe
df_angles_2
```
