```python
# these orientations have the direction heel->digitaltip
# For easier visualization of results I will invert the direction of the "vectors"
df_orientations_corrected = pd.DataFrame()
for column_name in df_orientations.columns:
    new_orientations=[]
    for orientation in df_orientations[column_name].values.astype(float):
        if orientation > 180:
            orientation-=180
            new_orientations.append(orientation)
        else:
            orientation+=180
            new_orientations.append(orientation)
    df_orientations_corrected[column_name] = new_orientations
# showing dataframe
df_orientations_corrected
```

```python
import math
# NEED TO BROADEN THE CODE FOR ALL ANGLES OF INTEREST (IT COMPUTES ONLY THE ANGLE_HINDPAWS)

# create an array containing the angle in each frame and then add the array to the dataframe as a column

# FIRST - create the array

# create zero array with the length of the dataframe
angle_array=np.zeros(len(df_orientations_corrected2))
for index in range(len(df_orientations_corrected2)):
    a = np.array([df_orientations_corrected2['x(cos(_))_LB'][index],df_orientations_corrected2['y(sin(_))_LB'][index]])
    b = np.array([df_orientations_corrected2['x(cos(_))_RB'][index],df_orientations_corrected2['y(sin(_))_RB'][index]])
    angle_array[index] = round(math.degrees(np.arccos(np.dot(a,b))),2)

angle_array
# SECOND - Add the array to the dataframe as a new column
df_orientations_corrected2['angle_hind_paws'] = angle_array.tolist()

# display the dataframe with the new column
df_orientations_corrected2
```
