```python
# create an array with the paw names
paws = ['left_frontlimb', 'right_frontlimb', 'left_hindlimb', 'right_hindlimb']

# creates an array with the paws from which the angle will be calculated
import itertools
combinations_paws = list(itertools.combinations(paws, 2))
print(combinations_paws)

# create a dictionary of the names for the angles of interest
angles_of_interest = {'angle_front_paws':combinations_paws[0], 
                      'angle_left_paws':combinations_paws[1],
                      'angle_leftfront_righthind':combinations_paws[2],
                      'angle_rightfront_lefthind':combinations_paws[3],
                      'angle_right_paws':combinations_paws[4],
                      'angle_hind_paws':combinations_paws[5]}
angles_of_interest
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import atan

# For loop that changes between the angle we want to obtain (iterates over the angles_of_interest dictionary)
for angle in angles_of_interest:
    # For loop that changes the values of x and y for the two points from which a line equation is obained
    # (for every angle, two lines will be obtained)
    index = 0
    # FIRST TWO POINTS - PAW1
    # Define the known points - digital tip
    # while index<len(df_video_raw)-1
    x_dt_h_1 = [float(df_video_raw[angles_of_interest[angle][0]+'_digitaltip (x)'][index]), float(df_video_raw[angles_of_interest[angle][0]+'_heel (x)'][index])]
    y_dt_h_1 = [float(df_video_raw[angles_of_interest[angle][0]+'_digitaltip (y)'][index]), float(df_video_raw[angles_of_interest[angle][0]+'_heel (y)'][index])]

    # FIRST TWO POINTS - PAW2
    # Define the known points -> x = [digital tip, heel], y = [digitaltip,heel]
    x_dt_h_2 = [float(df_video_raw[angles_of_interest[angle][1]+'_digitaltip (x)'][index]), float(df_video_raw[angles_of_interest[angle][1]+'_heel (x)'][index])]
    y_dt_h_2 = [float(df_video_raw[angles_of_interest[angle][1]+'_digitaltip (y)'][index]), float(df_video_raw[angles_of_interest[angle][1]+'_heel (y)'][index])]
    
    # check if the names and values are correct 
    print(angles_of_interest[angle][0])
    print(x_dt_h_1)
    print(y_dt_h_1)
    
    print(angles_of_interest[angle][1])
    print(x_dt_h_2)
    print(y_dt_h_2)
   
    # Calculates the coefficients (m and b, y = mx + b) for the two lines (one relative to paw1, and another relative to paw2)
    coefficients_paw1 = np.polyfit(x_dt_h_1, y_dt_h_1, 1)
    coefficients_paw2 = np.polyfit(x_dt_h_2, y_dt_h_2, 1)
    
    # Print the findings
    # m is the slope
    print('m_paw1 =', coefficients_paw1[0])
    # b is not important for defining the angle between two lines (only the slopes are)
    print('b_paw1 =', coefficients_paw1[1])
    # Print the findings
    # m is the slope
    print('m_paw2 =', coefficients_paw2[0])
    # b is not important for defining the angle between two lines (only the slopes are)
    print('b_paw2 =', coefficients_paw2[1])
        
    # Store the tan value  of the angle
    tan_angle = abs((coefficients_paw2[0] - coefficients_paw1[0]) / (1 + coefficients_paw1[0] * coefficients_paw2[0]))
 
    # Calculate tan inverse of the angle
    ret = atan(tan_angle)
 
    # Convert the angle from
    # radian to degree
    val = (ret * 180)/np.pi
 
    # Print the result
    print('The {} is {} degrees'.format(angle,round(val, 4)))
    
    print(' ')
```
