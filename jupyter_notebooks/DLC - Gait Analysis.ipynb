```python
# compute the euclidian distance between the nose and the tailbase
import math
euc_dist = math.sqrt((df_CM_calculation['nose (x)']-df_CM_calculation['tailbase (x)'])**2 + 
                     (df_CM_calculation['nose (y)']-df_CM_calculation['tailbase (y)'])**2)
print(euc_dist)
```
