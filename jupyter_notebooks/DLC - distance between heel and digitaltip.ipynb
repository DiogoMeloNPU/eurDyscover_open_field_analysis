## Distance between heel and digitaltip along the video

```python
df_predictions_raw
```

```python
x1, x2 = df_predictions_raw['right_hindlimb_heel (x)'].apply(lambda x: float(x)).to_numpy(), df_predictions_raw['right_hindlimb_digitaltip (x)'].apply(lambda x: float(x)).to_numpy()
y1, y2 = df_predictions_raw['right_hindlimb_heel (y)'].apply(lambda x: float(x)).to_numpy(), df_predictions_raw['right_hindlimb_digitaltip (y)'].apply(lambda x: float(x)).to_numpy()

x1_filtered, x2_filtered = df_predictions_filtered['right_hindlimb_heel (x)'].apply(lambda x: float(x)).to_numpy(), df_predictions_filtered['right_hindlimb_digitaltip (x)'].apply(lambda x: float(x)).to_numpy()
y1_filtered, y2_filtered = df_predictions_filtered['right_hindlimb_heel (y)'].apply(lambda x: float(x)).to_numpy(), df_predictions_filtered['right_hindlimb_digitaltip (y)'].apply(lambda x: float(x)).to_numpy()

x1_processed, x2_processed = df_processed['right_hindlimb_heel (x)'].apply(lambda x: float(x)).to_numpy(), df_processed['right_hindlimb_digitaltip (x)'].apply(lambda x: float(x)).to_numpy()
y1_processed, y2_processed = df_processed['right_hindlimb_heel (y)'].apply(lambda x: float(x)).to_numpy(), df_processed['right_hindlimb_digitaltip (y)'].apply(lambda x: float(x)).to_numpy()
```

```python
import math
dist_raw = []
for index in range(0,len(df_predictions_raw)):
    dist_raw.append(math.sqrt((x1[index]-x2[index])**2+(y1[index]-y2[index])**2))
```

```python
dist_filtered = []
for index in range(0,len(df_predictions_filtered)):
    dist_filtered.append(math.sqrt((x1_filtered[index]-x2_filtered[index])**2+(y1_filtered[index]-y2_filtered[index])**2))
```

```python
dist_processed = []
for index in range(0,len(df_processed)):
    dist_processed.append(math.sqrt((x1_processed[index]-x2_processed[index])**2+(y1_processed[index]-y2_processed[index])**2))
```

```python
# create a  new dataframe with the distances between heel and digitaltip (right paw)
# ... for both filtered and unfiltered
import math
df_predictions_raw['dist_right_paw'] = dist_raw
df_predictions_filtered['dist_right_paw'] = dist_filtered
df_processed['dist_right_paw'] = dist_processed
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib tk
plt.rcParams["figure.figsize"] = [12, 6]
plt.rcParams["figure.autolayout"] = True
#plot 1 - raw:
plt.subplot(3, 1, 1)
df_predictions_raw['dist_right_paw'].plot(kind='line')

#plot 2 - filtered:
plt.subplot(3, 1, 2)
df_predictions_filtered['dist_right_paw'].plot(kind='line', color = 'orange')

#plot 3 - processed:
plt.subplot(3,1,3)
df_processed['dist_right_paw'].plot(kind='line', color = 'k') # the peaks to '1000 px' disappear 

plt.show()
```
