```python
# this time: consider a movement initiation every time the velocity raises above 
# ... the beginning of a local minimum in the log distribution of velocities
# ... the local minimum is to be found in the valley forming the two peaks of
# ... the log_scale distribution

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

ax_nose = sns.kdeplot(df_diff_nose['V_xy_nose_SMA'], log_scale = True)
kde_curve = ax_nose.lines[0]
x = kde_curve.get_xdata()
y = kde_curve.get_ydata()
plt.plot(x, y, color='green')

peaks = np.where((y[1:-1] > y[0:-2]) * (y[1:-1] > y[2:]))[0] + 1
dips = np.where((y[1:-1] < y[0:-2]) * (y[1:-1] < y[2:]))[0] + 1

plt.plot (x, y)
plt.plot (x[peaks], y[peaks], 'o')
plt.plot (x[dips], y[dips], 'o')
plt.show()

print('This kde plot has 200 points!')
print('The local maxima are at the following positions (of the plot line!): {}'.format(peaks))
print('The local minima are at the following positions (of the plot line!): {}'.format(dips))

cutOff = x[dips[0]] #this value establishes the velocity treshold from which an initiation involving the right paw will be counted
print('Our cutOff for this mice will be: {}'.format(cutOff))
```
